```python
import set
import matplotlib.pyplot as plt
import tqdm

counter = {}

for i in tqdm.tqdm(range(10000)):

    count = set.CardCollection(set.Deck().draw(12)).set_count()

    counter[count] = counter.get(count, 0) + 1

for i in range(max(counter.keys()) + 1):
    counter[i] = counter.get(i, 0)
    print(f"{i}: {counter[i]}")

fig, ax = plt.subplots()

bars = ax.bar(counter.keys(), counter.values())
ax.bar_label(bars, padding=2)
plt.xticks(range(max(counter.keys()) + 1))

plt.title("Number of Sets in 10000 Random 12-card Collections")
plt.xlabel("Sets")
plt.ylabel("Collections")

plt.show()
```
